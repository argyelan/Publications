### Read in volume changes and EF ROI data

```python
ectdf = pd.read_csv('/resources/ANALYSIS/ECT_DATA_LABELS_2022-06-14_1917.csv')
efvol_t = pd.read_csv('/resources/ANALYSIS/Roast30-oldchanges.csv')
efvol_tt=efvol_t.iloc[np.where(efvol_t['volume change'].notnull())].copy()
amp = ectdf[['Record ID','ECT 1 amplitude (in mA, i.e. 800 or 900 typically)', 'ECT device', 'Age in years', 'Sex at birth of the participant']].iloc[np.where(ectdf['Event Name']=='baseline_before_treatment')]
amp.columns=['Record ID','amp', 'device', 'age', 'sex']
efvol = pd.merge(efvol_tt, amp, left_on='Record ID', right_on='Record ID')
efvol['amp2']=[(900 if (efvol.device[i]=='Thymatron') else 800) if np.isnan(value) else value for i,value in enumerate(efvol.amp)]
efvol['amp3']=[900 if value=='Thymatron' else 800 for i,value in enumerate(efvol.device)]
efvol['ef_mix2']=efvol['ef_mix']*(efvol['amp2']/900)
efvol['ef_mix3']=efvol['ef_mix']*(efvol['amp3']/900)
charges_df = pd.read_csv('/resources/ANALYSIS/charges249.csv')
#efvol
```

### Merging with clinical data

```python
def hue_regplot(data, x, y, hue, palette=None, **kwargs):
    from matplotlib.cm import get_cmap
    
    regplots = []
    
    levels = data[hue].unique()
    
    if palette is None:
        default_colors =  get_cmap('tab10')
        palette = {k: default_colors(i) for i, k in enumerate(levels)}
        
    for key in levels:
        regplots.append(
            sns.regplot(
                x = x,
                y = y,
                data = data[data[hue] == key],
                color = palette[key],
                **kwargs
            )
        )
    return regplots
```
